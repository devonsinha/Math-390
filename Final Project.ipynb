```python
fig = plt.figure(figsize = (10,10))
ax = plt.axes()
plt.style.use('fivethirtyeight')
def animate(year):
    ax.clear()
    ax.set_title("Years in the NBA: "+ str(year))
    ax.set_xlabel('PTS')
    ax.set_ylabel('PER')
    x = fulldataset.PTS[(fulldataset.career == year)]
    y = fulldataset.PER[(fulldataset.career == year)]
    ax.plot(x,y,'o', color = 'r', markersize = 10, alpha = 0.5)


sortedCareers=sorted(fulldataset.career.unique())
sortedCareers= [elem for elem in sortedCareers if elem > 0 and elem<24]

```

```python
ani = animation.FuncAnimation(fig,animate,sortedCareers, interval = 10000)
ani.save('NBA.gif', writer='devonsinha', fps=2)
filename = 'NBA.gif'
video = io.open(filename, 'r+b').read()
encoded = base64.b64encode(video)
HTML(data='''<img src="data:image/gif;base64,{0}" type="gif" />'''.format(encoded.decode('ascii')))
```
